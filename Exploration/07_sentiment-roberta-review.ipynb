```python
from google.colab import drive
drive.mount('/content/drive')
from transformers import pipeline, AutoTokenizer
import pickle
from tqdm import tqdm
from datasets import load_dataset
from transformers.pipelines.pt_utils import KeyDataset
import pandas as pd
```

```python
model_name = "siebert/sentiment-roberta-large-english"
sentiment_analysis = pipeline("sentiment-analysis",
                              model=model_name,
                              device=0,
                              batch_size=128,
                              truncation=True)
```

```python
dataset = load_dataset('csv',data_files='/content/drive/MyDrive/Colab Notebooks/a3case/Books_rating_sample.csv',split='train', usecols=['text'],)
dataset
```

```python
# L4 GPU batch_256 1500it [01:22, 18.18it/s]
# L4 GPU batch_128 15000it [13:09, 18.99it/s]
# T4 GPU batch_128 15000it [25:51,  9.67it/s]
```

```python
# L4 GPU batch_128
results = []
for out in tqdm(sentiment_analysis(KeyDataset(dataset, "text"))):
  results.append(out)
```

```python
with open("/content/drive/MyDrive/Colab Notebooks/a3case/Books_rating_sentiment_sample.pkl", 'wb') as file:
  pickle.dump(results, file)
```
